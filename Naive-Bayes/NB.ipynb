### Naive Bayes Machine Learning Algorithm Implementation

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,y=load_iris(return_X_y=True)
```

```python
print(X,y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,ConfusionMatrixDisplay
g=GaussianNB()
g.fit(X_train,y_train)
y_pred=g.predict(X_test)
print("Accuracy Score :",accuracy_score(y_pred,y_test))
print("Confusion Matrix :\n",confusion_matrix(y_pred,y_test))
print("classification_report : \n",classification_report(y_pred,y_test))
conf=confusion_matrix(y_pred,y_test)
cd=ConfusionMatrixDisplay(conf)
cd.plot()
```

```python
b=BernoulliNB()
b.fit(X_train,y_train)
y_pred=b.predict(X_test)
print("Accuracy Score :",accuracy_score(y_test,y_pred))
print("Confusion Matrix :\n",confusion_matrix(y_test,y_pred))
print("classification_report : \n",classification_report(y_test,y_pred))
conf1=confusion_matrix(y_test,y_pred)
cd1=ConfusionMatrixDisplay(conf1)
cd1.plot()
```
